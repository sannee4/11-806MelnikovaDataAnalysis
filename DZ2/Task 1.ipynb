```python
with_10_abs = data[(data['absences'] == 10)]
with_no_10_abs = data[(data['absences'] != 10)]
with_10_abs.shape, with_no_10_abs.shape
```

```python
with_10_abs_g3 = with_10_abs['G3'].mean()
with_no_10_abs_g3 = with_no_10_abs['G3'].mean()
print(with_10_abs_g3, with_no_10_abs_g3)
```

```python
have_paid = data[(data.paid == 'yes')]
do_not_have_paid = data[(data.paid == 'no')]
have_paid.shape, do_not_have_paid.shape
```

```python
have_paid_g3 = have_paid['G3'].mean()
do_not_have_paid_g3 = not_have_paid['G3'].mean()
print(have_paid_g3, do_not_have_paid_g3)
```

```python
have_paid_g3 = have_paid['G3'].mean()
do_not_have_paid_g3 = do_not_have_paid['G3'].mean()
print(have_paid_g3, do_not_have_paid_g3)
```

```python
plt.figure(figsize=(10,7))
plt.title('Traveltime to school')
data['traveltime'].hist()
plt.xlabel('sort of intervals')
plt.ylabel('number of students')
plt.show()
```

```python
travel_more_an_hour = data[(data['traveltime'] > 2)]
travel_more_an_hour.shape[0]/data.shape[0]*100
```
